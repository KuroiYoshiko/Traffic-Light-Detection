```python
import pandas as pd
import numpy as np
import seaborn as sns

from google.colab import files
uploaded = files.upload()
```

# **Analiza**

```python
results_x = pd.read_csv("results_x.csv", index_col=False)
results_l = pd.read_csv("results_l.csv", index_col=False)
results_m = pd.read_csv("results_m.csv", index_col=False)
results_s = pd.read_csv("results_s.csv", index_col=False)
results_n = pd.read_csv("results_n.csv", index_col=False)
edytowane_m = pd.read_csv("results_editedm.csv", index_col=False)

results_x.columns = results_x.columns.str.strip()
results_l.columns = results_l.columns.str.strip()
results_m.columns = results_m.columns.str.strip()
results_s.columns = results_s.columns.str.strip()
results_n.columns = results_n.columns.str.strip()
edytowane_m.columns = edytowane_m.columns.str.strip()

edytowane_m.head()
```

```python
results_n = results_n.drop(results_n[results_n['epoch'] > 500].index)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

g=sns.lineplot(data=results_x, x="epoch", y="metrics/precision(B)", label="YOLOv8x")
g=sns.lineplot(data=results_l, x="epoch", y="metrics/precision(B)", label="YOLOv8l")
g=sns.lineplot(data=results_m, x="epoch", y="metrics/precision(B)", label="YOLOv8m")
g=sns.lineplot(data=edytowane_m, x="epoch", y="metrics/precision(B)", label="YOLOv8m, imgsz=800")

sns.set(font_scale=1)

# Podpisanie ostatniego wyniku na wykresie
last_result = results_x.iloc[-1]['metrics/precision(B)']
plt.annotate(f'{last_result}', xy=(results_x['epoch'].iloc[-1], last_result),
             xytext=(results_x['epoch'].iloc[-1]-10, last_result+0.1),
             arrowprops=dict(arrowstyle='->', color='black'))

# Podpisanie ostatniego wyniku na wykresie
last_result = results_l.iloc[-1]['metrics/precision(B)']
plt.annotate(f'{last_result}', xy=(results_l['epoch'].iloc[-1], last_result),
             xytext=(results_l['epoch'].iloc[-1]-10, last_result+0.1),
             arrowprops=dict(arrowstyle='->', color='black'))

# Podpisanie ostatniego wyniku na wykresie
last_result = results_m.iloc[-1]['metrics/precision(B)']
plt.annotate(f'{last_result}', xy=(results_m['epoch'].iloc[-1], last_result),
             xytext=(results_m['epoch'].iloc[-1]-10, last_result+0.1),
             arrowprops=dict(arrowstyle='->', color='black'))


# Podpisanie ostatniego wyniku na wykresie
last_result = edytowane_m.iloc[22]['metrics/precision(B)']
plt.annotate(f'{last_result}', xy=(edytowane_m['epoch'].iloc[22], last_result),
             xytext=(edytowane_m['epoch'].iloc[22]-10, last_result+0.1),
             arrowprops=dict(arrowstyle='->', color='black'))

g.legend(loc='lower right')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

g=sns.lineplot(data=results_x, x="epoch", y="metrics/recall(B)")
g=sns.lineplot(data=results_l, x="epoch", y="metrics/recall(B)")
g=sns.lineplot(data=results_m, x="epoch", y="metrics/recall(B)")
g=sns.lineplot(data=results_s, x="epoch", y="metrics/recall(B)")

sns.set(font_scale=1)

# Podpisanie ostatniego wyniku na wykresie
last_result = results_x.iloc[-1]['metrics/recall(B)']
plt.annotate(f'x: {last_result}', xy=(results_x['epoch'].iloc[-1], last_result),
             xytext=(results_x['epoch'].iloc[-1]-10, last_result+0.1),
             arrowprops=dict(arrowstyle='->', color='black'))

# Podpisanie ostatniego wyniku na wykresie
last_result = results_l.iloc[-1]['metrics/recall(B)']
plt.annotate(f'l: {last_result}', xy=(results_l['epoch'].iloc[-1], last_result),
             xytext=(results_l['epoch'].iloc[-1]-10, last_result+0.1),
             arrowprops=dict(arrowstyle='->', color='black'))

# Podpisanie ostatniego wyniku na wykresie
last_result = results_m.iloc[-1]['metrics/recall(B)']
plt.annotate(f'm: {last_result}', xy=(results_m['epoch'].iloc[-1], last_result),
             xytext=(results_m['epoch'].iloc[-1]-10, last_result+0.1),
             arrowprops=dict(arrowstyle='->', color='black'))

# Podpisanie ostatniego wyniku na wykresie
last_result = results_s.iloc[-1]['metrics/recall(B)']
plt.annotate(f's: {last_result}', xy=(results_s['epoch'].iloc[-1], last_result),
             xytext=(results_s['epoch'].iloc[-1]-10, last_result+0.1),
             arrowprops=dict(arrowstyle='->', color='black'))

```
